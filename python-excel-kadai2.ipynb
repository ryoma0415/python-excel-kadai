```python
# import and setting
import pandas as pd
df = pd.DataFrame(data ={
    'Date': ['2023-05-17', '2023-05-18', '2023-05-19', '2023-05-20', '2023-05-21'],
    'Employee Name': ['Yamada', 'Satou', 'Suzuki', 'Tanaka', 'Takahashi'],
    'Earnings': [100, 200, 150, 300, 250],
    'Dpartment': ['Manufacturer', 'Agency', 'Manufacturer', 'Trading', 'Agency']
})

# set avarage column and int
df['average'] = df['Earnings'].mean()
average_int = df['Earnings'].mean()

# def gudge(performance is A, B, or C)
def gudge(earn):
  rank = '';
  if earn >= average_int + 50:
    rank = 'A';
  elif earn >= average_int:
    rank = 'B';
  else:
    rank = 'C'
  return rank

# set Performance column
df['Performance'] = df['Earnings'].apply(gudge)

# write
writer = pd.ExcelWriter('Performance.xlsx')

df.to_excel(writer, sheet_name='Sheet1', index=False)

writer.close()
```
